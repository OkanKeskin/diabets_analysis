```python
# Paketlerin yüklenmesi
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# csv tipi dosyanın okunması
path = "//content/diabetes.csv"
diabets = pd.read_csv(path)
```

```python
# veritabanı hakkında bilgi
diabets.info()
```

```python
# Boş verilerin yerine ortalama değer yazılması

diabets["Pregnancies"].fillna(diabets["Pregnancies"].mean(),inplace = True)
diabets["Glucose"].fillna(diabets["Glucose"].mean(),inplace = True)
diabets["BloodPressure"].fillna(diabets["BloodPressure"].mean(),inplace = True)
diabets["SkinThickness"].fillna(diabets["SkinThickness"].mean(),inplace = True)
diabets["Insulin"].fillna(diabets["Insulin"].mean(),inplace = True)
diabets["BMI"].fillna(diabets["BMI"].mean(),inplace = True)
diabets["DiabetesPedigreeFunction"].fillna(diabets["DiabetesPedigreeFunction"].mean(),inplace = True)
diabets["Age"].fillna(diabets["Age"].mean(),inplace = True)
diabets["Outcome"].fillna(diabets["Outcome"].mean(),inplace = True)
```

```python
 #Yinelenen satırları kaldırma
 
 diabets.drop_duplicates(inplace = True)
```

```python
# veritabanının sütunlarındaki değerlerin özellikleri
diabets.describe()
```

```python
#Yaşı ortalamadan büyük ve insülin değeri 30 dan küçük insanların verileri 
yas_insulin_veri = diabets[(diabets["Age"] > diabets["Age"].mean()) & (diabets["Insulin"] < 30)]

print(yas_insulin_veri)
```

```python
#Vücut kitle indeksi normal değer aralığında olan insanlar
yas_bmi_veri=diabets[(diabets["BMI"]>=18.5) & (diabets["BMI"]<=25)]
print(yas_bmi_veri)
```

```python
# Cilt kalınlığı 30'un üstünde, en fazla 3 çocuğu olan diyabet hastaları 
cilt30_cocuk_diyabet_veri=diabets[(diabets["SkinThickness"] > 30) & (diabets["Pregnancies"] <= 3) & (diabets["Outcome"]  == 1)]
print(cilt30_cocuk_diyabet_veri)
```

```python
#kan basıncı 70'in üzerinde ve glikoz değeri 100'ün üzerinde diyabet olmayan insanlar
kan70_glikoz100_not_diyabet_veri=diabets[(diabets["BloodPressure"] > 70) & (diabets["Glucose"] > 100) & (diabets["Outcome"]  == 0)]
print(kan70_glikoz100_not_diyabet_veri)
```

```python
#9 tane çocuğa sahip ve vücut kitle indeksi 32'nin üzerinde olan ve glikoz değeri 150'nin üzerinde olan 41 yaş üstü  diyabet hastaları
cocuk5_bmi30_diyabet_veri=diabets[(diabets["Pregnancies"] ==9) & (diabets["BMI"] > 32) & (diabets["Glucose"]>150)& (diabets["Age"]>41) & (diabets["Outcome"]  == 1)]
print(cocuk5_bmi30_diyabet_veri)
```

```python
#Diyabet olma olasılığı 0.5'in üzerinde ve glikoz oranı 120'nin üzerine olup diyabet olmayan 
diyabetOranı_glikoz_veri = diabets[(diabets["DiabetesPedigreeFunction"] > 0.5) & (diabets["Glucose"] > 120) & (diabets["Outcome"] == 0)]

print(diyabetOranı_glikoz_veri)
```

```python
#42 yaşında ve 1 çocuğa sahip ve bmi oranı 20'nin üzerinde olan ve cilt kalınlığı 20'nin üzerinde olan diyabet olan  hastalar
age42_cocuk5_bmi20_cilt20_diyabet_veri=diabets[(diabets["Age"] ==42) & (diabets["BMI"] > 20 ) & (diabets["Pregnancies"]==1)& (diabets["SkinThickness"]>20) & (diabets["Outcome"]  == 1)  ]
print(age42_cocuk5_bmi20_cilt20_diyabet_veri)
```

```python
#Diyabet olma oranı 0.5'in altında, 40 yaşından küçük ve glikoz oranı 110'un altında olup diyabet olan insanlar
diyabetOranı_yas_glikoz_diyabet_veri=diabets[(diabets["DiabetesPedigreeFunction"] < 0.5) & (diabets["Age"] < 40) & (diabets["Glucose"] < 110) & (diabets["Outcome"] == 1)]
print(diyabetOranı_yas_glikoz_diyabet_veri)
```

```python
#Diyabet olma oranı 0.8'in üzerinde, 50 yaşından büyük ve cilt kalınlığı 25'in altında olup diyabet olmayan insanlar
diyabetOranı_yas_cilt_diyabet_veri=diabets[(diabets["DiabetesPedigreeFunction"] > 0.8) & (diabets["SkinThickness"] < 25) & (diabets["Age"] > 50) & (diabets["Outcome"] == 0)]
print(diyabetOranı_yas_cilt_diyabet_veri)
```

```python
# Kan basıncı 70'in altında 18 yaşından büyük vucut kitle indexi normalden fazla ve hiç çocuğu olmayan diyabet hastaları
diyabetOranı_yas_glikoz_diyabet_veri=diabets[(diabets["BloodPressure"] < 70) & (diabets["Age"] > 18) & (diabets["BMI"] > 25) & (diabets["Pregnancies"] == 0) & (diabets["Outcome"] == 1)]
print(diyabetOranı_yas_glikoz_diyabet_veri)
```

```python
 # Kan basıncının yaşa göre saçılım grafiği
sns.lmplot(x='BloodPressure', y='Age', data=diabets, height = 10)
```

```python
# Diyabet genetik oranının diyabet olma ile oranı violin(keman) grafiğine göre dağılımı
diabets_color_palette = ['#78C850',
                    '#F08030',
                    '#6890F0',
                    '#A8B820',
                    '#A8A878',
                    '#A040A0',
                    '#F8D030',  
                    '#E0C068',  
                    '#EE99AC',  
                   ]
plt.figure(figsize=(15,8))
sns.violinplot(x = 'DiabetesPedigreeFunction', y = 'Outcome', data = diabets, palette=diabets_color_palette , height = 20)
```

```python
# Vücut kitle endeksinin çocuk sayısı ile oranını swarm(sürü) grafiği ile gösterilmesi
plt.figure(figsize=(15,8))
sns.swarmplot(x = 'BMI', y = 'Pregnancies', data = diabets, palette=diabets_color_palette)
```

```python
#Vüvut kitle indeksinin yaşa göre grafiği-YOĞUNLUK GRAFİĞİ
plt.figure(figsize=(15,8))
sns.kdeplot(diabets.Age, diabets.BMI)
```

```python
#Yaşa göre dağılımı veren Sütun Grafiği
plt.figure(figsize=(15,8))
sns.countplot(x='Age', data=diabets, palette=diabets_color_palette)


plt.xticks(rotation=-90)
```
